```python
import geoplot as gplt 
import geopandas as gpd 
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import mapclassify as mc
import geoplot.crs as gcrs
import numpy as np
import seaborn as sns
```

```python
#loading primary dataset 

bridge_points = gpd.read_file("../GennyPlots/ShapeFilesFile/NBI/National_Bridge_Inventory.shp")
pd.set_option('display.max_columns', None)
bridge_points.head()
```

```python
#loading in contextualizing data, county line polygons, MODOT District polygons, and MODOT maintained roads lines

county_bounds = gpd.read_file("../GennyPlots/ShapeFilesFile/MO_County_Boundaries/MO_County_Boundaries.shp")
county_bounds.to_crs(crs=None, epsg=4326, inplace=True)
districts = gpd.read_file("../GennyPlots/ShapeFilesFile/MO_MoDOT_Districts/MO_MoDOT_Districts.shp")
districts.to_crs(crs=None, epsg=4326, inplace=True)
roads = gpd.read_file("../GennyPlots/ShapeFilesFile/MO_MoDOT_Roads_Arcs/MO_MoDOT_Roads_Arcs.shp")
roads.to_crs(crs=None, epsg=4326, inplace=True)
```

```python
#finding top 5 highways in terms of total length of bridges on those highways
top_hwys = bridge_points[['STRUCTUR_3','FACILITY_C']].copy()
top_hwys['FACILITY_C'] = top_hwys['FACILITY_C'].apply(lambda x: x.strip(' N')).apply(lambda y: y.strip(' S')).apply(lambda z: z.strip(' E')).apply(lambda a: a.strip(' W'))
top_hwys = top_hwys.groupby('FACILITY_C').sum().sort_values(by='STRUCTUR_3', ascending=False).reset_index()
top_hwys.head(10)
```

```python
proj = gcrs.AlbersEqualArea(central_latitude=39.5055, central_longitude=-91.5238)

fig, ax = plt.subplots(1, figsize=(15, 15), subplot_kw={'projection': proj})
plt.tight_layout()

#adding in county lines at a low alpha to help orient viewer, but not overwhelm
gplt.polyplot(
    county_bounds,
    linewidth=.5,
    projection=proj,
    facecolor="none",
    edgecolor='black',
    alpha=0.5,
    ax=ax
)

#adding in MODOT districts to aid in contextualizing analysis
gplt.polyplot(
    districts,
    linewidth=2,
    projection=proj,
    facecolor="none",
    edgecolor='black',
    alpha=1,
    ax=ax,
)

gplt.pointplot(bridge_points.geometry,
               color="lightgrey", 
               s=1, 
               ax=ax, 
               alpha = .5,
               extent= districts.buffer(.1).total_bounds)
gplt.pointplot(bridge_points[(bridge_points['FACILITY_C']=="IS 70 E")|(bridge_points['FACILITY_C']=="IS 70 W")].geometry,
               color="midnightblue", 
               s=5, 
               ax=ax, 
               alpha = 1,
               label="IS 70",
               extent= districts.buffer(.1).total_bounds)
gplt.pointplot(bridge_points[((bridge_points['FACILITY_C']=="IS 44 E")|(bridge_points['FACILITY_C']=="IS 44 W"))].geometry,
               color="steelblue", 
               s=5, 
               ax=ax, 
               alpha = 1,
               label="IS 44",
               extent= districts.buffer(.1).total_bounds)
gplt.pointplot(bridge_points[(((bridge_points['FACILITY_C']=="IS 49 S")|(bridge_points['FACILITY_C']=="IS 49 N"))&(bridge_points['STRUCTURE_']!='4152'))].geometry,
               color="teal", 
               s=5, 
               ax=ax, 
               alpha = 1,
               label="IS 49",
               extent= districts.buffer(.1).total_bounds)
gplt.pointplot(bridge_points[((bridge_points['FACILITY_C']=="IS 55 S")|(bridge_points['FACILITY_C']=="IS 55 N"))].geometry,
               color="seagreen", 
               s=5, 
               ax=ax, 
               alpha = 1,
               label="IS 55",
               extent= districts.buffer(.1).total_bounds)
gplt.pointplot(bridge_points[((bridge_points['FACILITY_C']=="IS 435 E")|(bridge_points['FACILITY_C']=="IS 435 W")|(bridge_points['FACILITY_C']=="IS 435 N")|(bridge_points['FACILITY_C']=="IS 435 S"))].geometry,
               color="yellowgreen", 
               s=5, 
               ax=ax, 
               alpha = 1,
               label="IS 435",
               extent= districts.buffer(.1).total_bounds)
ax.legend(fontsize=20)
ax.set_title("Top 5 Highways by Bridge Miles", fontsize=20);
fig.savefig("fig4.png")
```
